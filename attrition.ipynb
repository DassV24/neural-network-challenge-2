```python
                                                                # Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from sklearn.preprocessing import OneHotEncoder

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# View the types of data
attrition_df.dtypes
```

```python
# Create a list of at least 10 column names to use as X data
X_columns = ['Age', 'HourlyRate', 'DistanceFromHome', 'Education', 'WorkLifeBalance', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction']


# Create X_df using your selected columns
X_df = attrition_df[X_columns]

# Show the data types for X_df
X_df.head()
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# X_train = X_train.apply(pd.to_numeric, errors='coerce')
# X_test = X_test.apply(pd.to_numeric, errors='coerce')
# Add new code cells as necessary
```

```python
# Create a StandardScaler
scaler = StandardScaler().fit(X_train)

# Fit the StandardScaler to the training data
X_train_scaled = scaler.transform(X_train)

# Scale the training and testing data
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
department_onehot = OneHotEncoder(drop=None, handle_unknown='ignore', sparse_output=False)

# Fit the encoder to the training data
department_onehot.fit(y_train['Department'].values.reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_dpt_encoded = department_onehot.transform(np.array(y_train['Department']).reshape(-1, 1))
y_test_dpt_encoded = department_onehot.transform(np.array(y_test['Department']).reshape(-1, 1))

print(f"y_train_dpt_encoded: \n {y_train_dpt_encoded[:5]} \ny_test_dpt_encoded: \n {y_test_dpt_encoded[:5]}")
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_onehot = OneHotEncoder(drop=None, handle_unknown='ignore', sparse_output=False)

# Fit the encoder to the training data
attrition_onehot.fit(y_train['Attrition'].values.reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_att_encoded = attrition_onehot.transform(np.array(y_train['Attrition']).reshape(-1, 1))
y_test_att_encoded = attrition_onehot.transform(np.array(y_test['Attrition']).reshape(-1, 1))

# Print shape of encoded results
print(f"y_train_att_encoded: \n {y_train_att_encoded[:5]} \ny_test_att_encoded: \n {y_test_att_encoded[:5]}")
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_features = X_train_scaled.shape[1]
print(input_features)

# Create the input layer
input_layer = layers.Input(shape=(input_features,))

# Create at least two shared layers
shared_layer_1 = layers.Dense(units=64, activation='relu')(input_layer)
print(shared_layer_1.shape)
shared_layer_2 = layers.Dense(units=32, activation='relu')(shared_layer_1)
print(shared_layer_2.shape)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dpt_hidden_layer = layers.Dense(units=16, activation='relu')(shared_layer_2)
print(dpt_hidden_layer)
# Create the output layer
dpt_output_layer = layers.Dense(units=y_train_dpt_encoded.shape[1], activation='softmax', name='Department_Output')(dpt_hidden_layer)
print(dpt_output_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
att_hidden_layer = layers.Dense(units=16, activation='relu')(shared_layer_2)
print(att_hidden_layer)
# Create the output layer
att_output_layer = layers.Dense(units=y_train_att_encoded.shape[1], activation='softmax', name='Attrition_Output')(att_hidden_layer)
print(att_output_layer)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dpt_output_layer, att_output_layer])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics={'Department_Output':'accuracy','Attrition_Output':'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
train_model = model.fit(
    X_train_scaled,
    [y_train_dpt_encoded, y_train_att_encoded],
    epochs=100,
    batch_size=32,
    validation_data=(X_test_scaled, [y_test_dpt_encoded, y_test_att_encoded])
    )
```

```python
# Accuracy for both Department_Output and Attrition_Output format for readability
# Evaluate the model with the testing data
results = model.evaluate(X_test_scaled, [y_test_dpt_encoded, y_test_att_encoded])
```

```python
# Print the accuracy for both department and attrition
print(f"Total Loss: {results[0]:.4f}")
print(f"Department Loss: {results[1]:.4f}\nDepartment Accuracy: {results[3]:.4f}")
print(f"Attrition Loss: {results[2]:.4f}\nAttrition Accuracy: {results[4]:.4f}")
```
